так как кол-во строк осталось равным 3276, то можно сделать вывод, что дубликатов в базе нет

```python
data.isnull()
```

```python
data_clear=data.dropna(subset=['Trihalomethanes'])
```

```python
data_clear['ph']=data_clear['ph'].fillna(0)
data_clear['Sulfate']=data_clear['Sulfate'].fillna(data_clear['Sulfate'].mean())
```

решила заменить в столбце ph пустые значения на 0, а в столбце suifate на сренее, т.к. думаю значение первого столбца категорично важно для точности анализа по данной выборке

```python
data_clear.info()
data_clear.describe()
```

можно выполнять нормализацию

```python
normalized_data_clear=normalize(data_clear, norm='l1', axis=1)
normalized_df=pd.DataFrame(normalized_data_clear,columns=data_clear.columns)
print(normalized_df)
```
